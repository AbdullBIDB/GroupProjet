```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# File to Load (Remember to Change These)
data_to_load = "fastest_growth_occupations.csv"

data_to_load
```

```python
original = pd.read_csv(data_to_load)
original.head()
```

```python
df_original = pd.DataFrame(original)
```

```python
df_original.columns = ['Occupations','ID','2016','2026','total_chage','Percent_change', 'Median annual wage']
df_original.head()
```

```python
cleaned = df_original.dropna(how = "any")
cleaned.reset_index(drop=True)
```

```python
tech_occupation_growth = cleaned.loc[(cleaned["Occupations"] == 'Information security analysts') | \
                                    (cleaned["Occupations"] == 'Software developers, applications') | \
                                    (cleaned["Occupations"] == 'Wind turbine service technicians') | \
                                    (cleaned["Occupations"] == 'Statisticians') | \
                                    (cleaned["Occupations"] == 'Operations research analysts')]
tech_occupation_growth
```

```python
tech_occupation_growth_2016 = tech_occupation_growth["2016"].astype(float)
```

```python
#tech_occupation_growth["Percent_change"]=tech_occupation_growth["Percent_change"].astype(float)
```

```python
#tech_occupation_growth.dtypes
```

```python
tech_2016=tech_occupation_growth["2016"].sum()
tech_2016
```

```python
tech_occupation_growth_2026 = tech_occupation_growth["2026"].str.replace(",","")
```

```python
tech_occupation_growth_2026 = tech_occupation_growth_2026.astype(float)
tech_occupation_growth_2026
```

```python
tech_2026=tech_occupation_growth_2026.sum()
tech_2026
```

```python
Tech_Compare_DataFrame = pd.DataFrame([tech_2016,tech_2026], ["2016", "2026"])
Tech_Compare_DataFrame.head()
```

```python
PerctChngeTech=Tech_Compare_DataFrame.pct_change()
```

```python

Tech_Compare_DataFrame.plot(kind="bar", legend=False)
plt.title("Expected_Tech_Growth")
plt.xlabel("Year")
plt.ylabel("Total x1000")
plt.savefig("./BarTech.jpg")
plt.show()
```

```python
searchfor = ['Information security analysts','Software developers, applications', 'Wind turbine service technicians','Statisticians','Operations research analysts']
other_occupations_growth = cleaned[~cleaned["Occupations"].str.contains('|'.join(searchfor))]
other_occupations_growth = other_occupations_growth.reset_index()
other_occupations_growth= other_occupations_growth.drop("index", 1)
other_occupations_growth
```

```python
other_occupations_growth['2016']=other_occupations_growth['2016'].str.replace(",","")
other_occupations_growth['2026']=other_occupations_growth['2026'].str.replace(",","")
other_occupations_growth=other_occupations_growth.drop([0,0])
#restocc=float(restocc)
#print(restocc)
other_occupations_growth
```

```python
other_occupations_growth["2016"]=other_occupations_growth["2016"].astype(float)
other_occupations_growth["2026"]=other_occupations_growth["2026"].astype(float)
```

```python
other_2016=sum(other_occupations_growth["2016"])
other_2016
```

```python
other_2026 =sum(other_occupations_growth["2026"])
other_2026
```

```python
Other_Compare_DataFrame = pd.DataFrame([other_2016, other_2026], ["2016", "2026"])
Other_Compare_DataFrame.head()
```

```python
PerctChngeOther=Other_Compare_DataFrame.pct_change()
PerctChngeOther
```

```python
Other_Compare_DataFrame.plot(kind="bar", legend=False)
plt.title("Expected_Other_Growth")
plt.xlabel("Year")
plt.ylabel("Total x1000")
plt.savefig("./BarOther.jpg")
plt.show()
```
