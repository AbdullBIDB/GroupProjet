```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


```

```python
# File to Load (Remember to change these)
wage_research_2017= "natsector_2017.csv"
wage_research_2007 = "natsector_2007.csv"
wage_research_2012 = "natsector_2012.csv"
wage_research_2008= "natsector_2008.csv"
wage_research_2009= "natsector_2009.csv"
wage_research_2010 = "natsector_2010.csv"
wage_research_2011 = "natsector_2011.csv"
wage_research_2013 = "natsector_2013.csv"
wage_research_2014 = "natsector_2014.csv"
wage_research_2015 = "natsector_2015.csv"
wage_research_2016 = "natsector_2016.csv"


# Read the City and Ride Data
wages_2007 = pd.read_csv(wage_research_2007,index_col=[0])
wages_2017 = pd.read_csv(wage_research_2017,index_col=[0])
wages_2012 = pd.read_csv(wage_research_2012,index_col=[0])
wages_2008 = pd.read_csv(wage_research_2008,index_col=[0])
wages_2009 = pd.read_csv(wage_research_2009,index_col=[0])
wages_2010 = pd.read_csv(wage_research_2010,index_col=[0])
wages_2011 = pd.read_csv(wage_research_2011,index_col=[0])
wages_2013 = pd.read_csv(wage_research_2013,index_col=[0])
wages_2014 = pd.read_csv(wage_research_2014,index_col=[0])
wages_2015 = pd.read_csv(wage_research_2015,index_col=[0])
wages_2016 = pd.read_csv(wage_research_2016,index_col=[0])

#wages_2007_df = pd.DataFrame(wages_2007)


#wages_2007_df= wages_2007_df.groupby("NAICS_TITLE")
#wages_2007_df.head()

```

```python
combined_data = pd.concat([wages_2007, wages_2008, wages_2009, 
                         wages_2010, wages_2011, wages_2012, 
                         wages_2013, wages_2014, wages_2015, 
                           wages_2016, wages_2017], axis=1, join='outer').sort_index()
```

```python

#combined_data["2007 A_MEAN"] = combined_data["2007 A_MEAN"].apply(pd.to_numeric, errors='coerce')
combined_data["2007 A_MEAN"] = pd.to_numeric(combined_data["2007 A_MEAN"], errors='raise')
combined_data["2008 A_MEAN"] = pd.to_numeric(combined_data["2008 A_MEAN"], errors='raise')
combined_data["2009 A_MEAN"] = pd.to_numeric(combined_data["2009 A_MEAN"], errors='raise')
combined_data["2010 A_MEAN"] = pd.to_numeric(combined_data["2010 A_MEAN"], errors='raise')
combined_data["2011 A_MEAN"] = pd.to_numeric(combined_data["2011 A_MEAN"], errors='raise')
combined_data["2012 A_MEAN"] = pd.to_numeric(combined_data["2012 A_MEAN"], errors='raise')
combined_data["2013 A_MEAN"] = pd.to_numeric(combined_data["2013 A_MEAN"], errors='raise')
combined_data["2014 A_MEAN"] = pd.to_numeric(combined_data["2014 A_MEAN"], errors='raise')
combined_data["2015 A_MEAN"] = pd.to_numeric(combined_data["2015 A_MEAN"], errors='raise')
combined_data["2016 A_MEAN"] = pd.to_numeric(combined_data["2016 A_MEAN"], errors='raise')
combined_data["2017 A_MEAN"] = pd.to_numeric(combined_data["2017 A_MEAN"], errors='raise')
```

```python
#obtain the x and y coordinates 

Manufacturing = combined_data.loc["Manufacturing",["2008 A_MEAN","2009 A_MEAN","2010 A_MEAN", "2011 A_MEAN", "2012 A_MEAN", "2013 A_MEAN", "2014 A_MEAN", "2015 A_MEAN", "2016 A_MEAN","2017 A_MEAN"]]
years= ['08', '09', '10', '11', '12', '13', '14', '15', '16','17']
Transportation = combined_data.loc["Transportation and Warehousing",["2008 A_MEAN","2009 A_MEAN","2010 A_MEAN", "2011 A_MEAN", "2012 A_MEAN", "2013 A_MEAN", "2014 A_MEAN", "2015 A_MEAN", "2016 A_MEAN","2017 A_MEAN"]]
Health_Care = combined_data.loc["Health Care and Social Assistance",["2008 A_MEAN","2009 A_MEAN","2010 A_MEAN", "2011 A_MEAN", "2012 A_MEAN", "2013 A_MEAN", "2014 A_MEAN", "2015 A_MEAN", "2016 A_MEAN","2017 A_MEAN"]]
Construction = combined_data.loc["Construction",["2008 A_MEAN","2009 A_MEAN","2010 A_MEAN", "2011 A_MEAN", "2012 A_MEAN", "2013 A_MEAN", "2014 A_MEAN", "2015 A_MEAN", "2016 A_MEAN","2017 A_MEAN"]]
Education = combined_data.loc["Educational Services",["2008 A_MEAN","2009 A_MEAN","2010 A_MEAN", "2011 A_MEAN", "2012 A_MEAN", "2013 A_MEAN", "2014 A_MEAN", "2015 A_MEAN", "2016 A_MEAN","2017 A_MEAN"]]

#plot annual average wage per year

#plt.plot(years, Construction, color="orange", label="Construction")
#plt.plot(years, Health_Care, color='green', label="Health Care")
plt.plot(years, Manufacturing, color="blue", label="Manufacturing")
plt.plot(years, Transportation, color ='red', label="Transportation")
#plt.plot(years,Education, color = 'purple', label="Education")

#plot details
plt.title("Growth of Average Wages 2008-2017")
plt.ylabel("Salary")
plt.xlabel("Year 20-")
plt.grid()

plt.legend(fontsize="medium", loc='best', mode="expanded", labelspacing=0.5)
plt.savefig("manufacturing_transportation.png")
```

```python
#compare mean to median

Manufacturing_Median = combined_data.loc["Manufacturing",["2008 A_MEDIAN","2009 A_MEDIAN","2010 A_MEDIAN", "2011 A_MEDIAN", "2012 A_MEDIAN", "2013 A_MEDIAN", "2014 A_MEDIAN", "2015 A_MEDIAN", "2016 A_MEDIAN","2017 A_MEDIAN"]]
plt.plot(years, Manufacturing, color="blue", label="Mean")
plt.plot(years, Manufacturing_Median, color="red", label="Median")

#plot details
plt.title("Manufacturing Mean : Median Wages 2008-2017")
plt.ylabel("Annual Salary")
plt.xlabel("Year 20-")
plt.grid()

plt.legend(fontsize="medium", loc='best', mode="expanded", labelspacing=0.5)
plt.savefig("mean_median_wages.png")
```
