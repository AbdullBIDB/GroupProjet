```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

bls_prod_data = "Resources/Manufacturing_Prod_Costs.csv"
bls_labor_data = "Resources/Manufacturing_Labor_Costs.csv"
automation_data = "Resources/Automation_Labor_Costs.csv"

# Read data
annual_prod = pd.read_csv(bls_prod_data)
annual_labor = pd.read_csv(bls_labor_data)
automation_labor_costs = pd.read_csv(automation_data)
```

```python
annual_prod
```

```python
annual_labor
```

```python
years = annual_prod['Year']
prod_values = annual_prod['Labor Productivity % Change']
labor_values = annual_labor['Labor Costs % Change']
plt.plot(years, prod_values, color = 'blue', linewidth = 2, linestyle = "-")
plt.plot(years, labor_values, color = 'red', linewidth = 2, linestyle = "--")
plt.title("Manufacturing Labor Productivity and Costs")
plt.xlabel("Year")
plt.ylabel("Labor Productivity and Cost Index")
plt.ylim(-10.0,10.0)
plt.xlim(2007,2017)
plt.legend(loc = 'upper left');
plt.grid(True)
plt.tight_layout()
plt.savefig('Manufacturing Labor Productivity and Costs.jpg')
plt.show()
```

```python
years = automation_labor_costs['Year']
labor_values = automation_labor_costs['Labor Costs']
robot_values = automation_labor_costs['Robot Prices']
plt.bar(years, labor_values, color="navy", align="center", label="Labor Costs")
plt.bar(years, robot_values, color="orange", align="center", label="Robot Prices")
plt.title("Labor Costs vs Robot Prices")
plt.xlabel("Year")
plt.ylabel("Labor Cost and Robot Price Index")
plt.legend();
plt.tight_layout()
plt.savefig('Labor Costs vs Robot Costs.jpg')
plt.show()
```
